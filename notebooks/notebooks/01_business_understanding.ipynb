```python
# ============================================================
# BUSINESS UNDERSTANDING
# Project: Credit Risk & Loan Portfolio Analytics for an NBFC
# ============================================================

# What is a Loan?
# - A loan is a financial agreement in which a lender provides funds to a borrower.
# - The borrower agrees to repay the principal amount along with interest over a specified period.
# - Loans are typically repaid through fixed monthly installments (EMIs).
# - Financial institutions assess a borrower's creditworthiness before approving a loan.

# What is Loan Default?
# - Loan default occurs when a borrower fails to make scheduled repayments as agreed.
# - A loan is considered defaulted after a prolonged period of missed payments.
# - Defaults lead to financial losses for the lender and increase the overall portfolio risk.
# - Predicting potential defaults helps financial institutions minimize credit losses.

# Why do Customers Default?
# - Low or unstable income
# - High debt-to-income (DTI) ratio
# - Poor credit history or low credit score
# - High interest rates resulting in larger EMIs
# - Long loan tenure
# - Unexpected financial emergencies such as job loss or medical expenses
# - Excessive existing debt obligations

# Which Customers are Risky?
# - Customers with low annual income
# - Customers with high debt-to-income ratios
# - Customers with poor credit grades (e.g., Grade D-G)
# - Customers with low FICO scores
# - Customers with high credit utilization
# - Customers with previous delinquencies
# - Customers requesting large loan amounts relative to their income

# Business Objective
# - Analyze the loan portfolio to identify high-risk borrowers.
# - Understand the key factors contributing to loan defaults.
# - Build an analytics solution to monitor portfolio health.
# - Develop a machine learning model to predict loan defaults.
# - Provide actionable business recommendations to improve lending decisions.

# Key Performance Indicators (KPIs)
# - Total Loan Applications
# - Total Funded Amount
# - Average Loan Amount
# - Average Interest Rate
# - Average Annual Income
# - Default Rate (%)
# - Fully Paid Loan Rate (%)
# - Default Rate by Loan Grade
# - Default Rate by Loan Purpose
# - Default Rate by State
# - Default Rate by Income Bracket
# - Average Debt-to-Income Ratio (DTI)
# - Average FICO Score
# - Portfolio Distribution by Loan Term (36 vs 60 months)
# - Portfolio Distribution by Home Ownership
```
